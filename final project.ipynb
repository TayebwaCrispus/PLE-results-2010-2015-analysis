### Normalizing data types.
Some of the columns forexample 'FEMALE CANDIDATES' and 'MALE CANDIDATES' are supposed to be integer data types but by default were set to be floats.
Inorder to convert the data type from float to int in the 'MALE CANDIDATES' and 'FEMALE CANDIDATES' we are going to fill in the missing values as 0

```python
df['MALE CANDIDATES'] = df['MALE CANDIDATES'].fillna(0)
df['FEMALE CANDIDATES'] = df['FEMALE CANDIDATES'].fillna(0)
df['FEMALE CANDIDATES'] = df['FEMALE CANDIDATES'].astype(int)
df['MALE CANDIDATES'] = df['MALE CANDIDATES'].astype(int)
```

As viewed above, all datatypes are normalized.

### Calculating the total number of candidates per year

```python
CandidatesPerYear = df[['YEAR', 'SCHOOL', 'TOTAL CANDIDATES']]
CandidatesPerYear
```

```python
def highlight_max(j):
    is_max = j == j.max()
    return ['background-color: green' if cell else '' for cell in is_max]
def highlight_min(j):
    is_min = j == j.min()
    return ['background-color: yellow' if cell else '' for cell in is_min]
totalCandidatesPerYear = CandidatesPerYear.groupby('YEAR').sum()
totalCandidatesPerYear.style.apply(highlight_max)
```

```python
CandidatesPerYearbarh = totalCandidatesPerYear.plot( kind = 'bar', align='center', figsize=(13,5), color = 'red', legend=False,
                         title = "GRAPH OF THE TOTAL NUMBER OF CANDIDATES WHO SIT FOR THE PLE EXAMS",
                        ).set_xlabel('')
```

Analyzing the graph above, 2013 had the most candidates.
